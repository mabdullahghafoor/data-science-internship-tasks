The objective of this task is to explore and visualize the Iris dataset using Python libraries such as pandas, matplotlib, and seaborn.

This notebook includes:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.hist(iris["sepal_length"], bins=20)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(data=iris)

plt.title("Box Plot of Iris Features")
plt.show()
```

## Observations

- The dataset contains 150 flower samples.
- There are 5 columns including the species column.
- No missing values were found in the dataset.
- The species are evenly distributed.
- The scatter plot shows relationships between sepal length and petal length.
- The histogram shows the distribution of sepal length.
- The box plot helps identify spread and possible outliers.
